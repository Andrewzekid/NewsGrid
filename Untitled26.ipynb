```python
def make_discriminator_model():
  model = tf.keras.Sequential()
  model.add(keras.layers.Conv2D(64,kernel_size=(4,4),padding="same",activation="relu",input_shape=(28,28,1)))
  model.add(keras.layers.MaxPool2D())
  model.add(keras.layers.Conv2D(32,kernel_size=(4,4),padding="same",activation="relu"))
  model.add(keras.layers.MaxPool2D())
  model.add(keras.layers.Conv2D(16,kernel_size=(4,4),padding="same",activation="relu"))
  model.add(keras.layers.Flatten())
  model.add(keras.layers.Dense(1,activation="sigmoid"))
  return model
```

```python
def make_generator_model():
  model = tf.keras.Sequential()
  model.add(tf.keras.layers.Dense(256*7*7,input_shape=(100,)))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.Reshape((7,7,256)))
  
  model.add(tf.keras.layers.Conv2DTranspose(64,strides=(2,2),kernel_size=(4,4),padding="same"))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.LeakyReLU(alpha=0.2))

  model.add(tf.keras.layers.Conv2DTranspose(32,strides=(2,2),kernel_size=(4,4),padding="same"))
  model.add(tf.keras.layers.BatchNormalization())
  model.add(tf.keras.layers.LeakyReLU(alpha=0.2))

  model.add(tf.keras.layers.Dropout(0.5))
  model.add(tf.keras.layers.Conv2DTranspose(1,kernel_size=(4,4),activation="tanh",padding="same"))
  return model

```

```python
generator = make_generator_model()
```

```python
def generator_loss(fake_predictions):
  fake_loss = tf.losses.binary_crossentropy(tf.ones_like(fake_predictions),fake_predictions)
  return fake_loss
```

```python
def discriminator_loss(fake_predictions,real_predictions):
  fake_loss = tf.losses.binary_crossentropy(tf.zeros_like(fake_predictions),fake_predictions)
  real_loss = tf.losses.binary_crossentropy(tf.ones_like(real_predictions),real_predictions)
  return fake_loss + real_loss
```

```python
generator_optimizer = tf.keras.optimizers.Adam(1e-3)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-3)
```

```python
noise_dim=100
epochs=50
```

```python
BATCH_SIZE=256
```

```python
@tf.function
def train_step(images):
  noise = np.random.randn([BATCH_SIZE,100])
  with tf.GradientTape() as gen_tape,tf.GradientTape() as disc_tape:
    generated_images = generator(noise)
    real_predictions = discriminator(images)
    fake_predictions = discriminator(generated_images)

    gen_loss = generator_loss(fake_predictions)
    disc_loss = discriminator_loss(fake_predictions,real_predictions)

  generator_gradients = gen_tape(gen_loss,generator.trainable_variables)
  discriminator_gradients = disc_tape(disc_loss,discriminator.trainable_variables)
  
  generator_optimizer.apply_gradients(zip(generator_gradients,generator.trainable_variables))
  discriminator_optimizer.apply_gradients(zip(discriminator_gradients,discriminator.trainable_variables))
  
```

```python
@tf.function
def train(dataset,epochs):
  for epoch in range(epochs):
    for image_batch in dataset:
      train_step(image_batch)
```

```python
train(dataset,epochs)
```
